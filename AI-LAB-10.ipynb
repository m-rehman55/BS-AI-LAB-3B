```python
import pandas as pd
import numpy as np

# --- Lab 9 se continue: Data Load karna ---
data_set = pd.read_csv('CrudeOil_Data.csv')

# --- Lab 10: Data Pre-Processing ---

# Checking null values
np.sum(pd.isnull(data_set))

# Filling null values (Example)
# NOTE: 'Specific_Null_Column' ko apne data ke column naam se badal dein
column_name_to_fill = 'Specific_Null_Column'

if column_name_to_fill in data_set.columns and data_set[column_name_to_fill].isnull().any():
    num = data_set[column_name_to_fill].mode()[0]
    data_set[column_name_to_fill] = data_set[column_name_to_fill].fillna(num)

# Dropping unnecessary columns (Example)
# NOTE: 'id' ko apni actual unnecessary column se badal dein.
column_to_drop = 'id'

if column_to_drop in data_set.columns:
    data_set.drop(column_to_drop, axis=1, inplace=True)

# Check datatypes of all columns
data_set.dtypes

# Splitting into X (Features) and Y (Target)
X = data_set.iloc[:, 0:-1]
X.shape

Y = data_set.iloc[:,-1]
Y.shape

# Converting Object columns into Int columns
cat_columns = X.select_dtypes(['object']).columns

if len(cat_columns) > 0:
    X[cat_columns] = X[cat_columns].apply(lambda x: pd.factorize(x)[0])
```
